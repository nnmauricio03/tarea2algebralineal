#Ejercicio 4. ¿Cómo evalúa el módulo NAcAL las siguientes expresiones?


$(a) 3*b|1$


$(b) 3*1|b$


$(c) 1*3|b$


$(d) 1|b*3$

```python
b = Vector( [0,4,6] )
print("(a)")
display(Latex(latex(3*b|1)))
print("(b)")
display(Latex(latex(3*1|b)))
print("(c)")
display(Latex(latex(1*3|b)))
print("(d)")
display(Latex(latex(1|b*3)))
```

#Ejercicio 5. Demuestre las propiedades de la siguiente proposición.

**1. A + B = B + A**

```python
a = Vector( [1,3,5] )
b = Vector( [7,9,11] )
c = Vector( [13,15,17] )
A = Matrix( [a,b,c] )
print("Matrix A")
display(Latex(latex(A)))
print()
d = Vector( [2,4,6] )
e = Vector( [8,10,12] )
f = Vector( [14,16,18] )
B = Matrix( [d,e,f] )
print("Matrix B")
display(Latex(latex(B)))
print()
print()
print("Matrix A + B")
display(Latex(latex(A+B)))
print()
print("Matrix B + A")
display(Latex(latex(B+A)))
```

**2. A + (B + C) = (A + B) + C**

```python
a = Vector( [1,3,5] )
b = Vector( [7,9,11] )
c = Vector( [13,15,17] )
A = Matrix( [a,b,c] )
print("Matrix A")
display(Latex(latex(A)))
print()
d = Vector( [2,4,6] )
e = Vector( [8,10,12] )
f = Vector( [14,16,18] )
B = Matrix( [d,e,f] )
print("Matrix B")
display(Latex(latex(B)))
print()
g = Vector( [1,2,3] )
h = Vector( [4,5,6] )
i = Vector( [7,8,9] )
C = Matrix( [g,h,i] )
print("Matrix C")
display(Latex(latex(C)))
print()
print()
print("Matrix A + (B + C)")
display(Latex(latex(A+(B+C))))
print()
print("Matrix (A + B) + C")
display(Latex(latex((A+B)+C)))
```

**3. 0 + A = A**

```python
a = Vector( [0,0,0] )
b = Vector( [0,0,0] )
c = Vector( [0,0,0] )
A = Matrix( [a,b,c] )
print("Matrix 0")
display(Latex(latex(A)))
print()
d = Vector( [2,4,6] )
e = Vector( [8,10,12] )
f = Vector( [14,16,18] )
B = Matrix( [d,e,f] )
print("Matrix A")
display(Latex(latex(B)))
print()
print()
print("Matrix 0 + A = A")
display(Latex(latex(A+B)))
```

**4. A + (−A) = 0.**

```python
a = Vector( [2,4,6] )
b = Vector( [8,10,12] )
c = Vector( [14,16,18] )
A = Matrix( [a,b,c] )
print("Matrix A")
display(Latex(latex(A)))
print()
d = Vector( [-2,-4,-6] )
e = Vector( [-8,-10,-12] )
f = Vector( [-14,-16,-18] )
B = Matrix( [d,e,f] )
print("Matrix -A")
display(Latex(latex(B)))
print()
print()
print("Matrix A + (-A)")
display(Latex(latex(A+B)))
```

**5. λ(A + B) = λA + λB**

```python
a = Vector( [1,3,5] )
b = Vector( [7,9,11] )
c = Vector( [13,15,17] )
A = Matrix( [a,b,c] )
print("Matrix A")
display(Latex(latex(A)))
print()
d = Vector( [2,4,6] )
e = Vector( [8,10,12] )
f = Vector( [14,16,18] )
B = Matrix( [d,e,f] )
print("Matrix B")
display(Latex(latex(B)))
print()
print()
print("λ=2")
print("Matrix λ(A + B)")
display(Latex(latex(2*(A+B))))
print()
print("Matrix λB + λA")
display(Latex(latex((2*A)+(2*B))))
```

**6. (λ + η)A = λA + ηA**

```python
a = Vector( [1,3,5] )
b = Vector( [7,9,11] )
c = Vector( [13,15,17] )
A = Matrix( [a,b,c] )
print("Matrix A")
display(Latex(latex(A)))
print()
print()
print("λ=2  η=3")
print("Matrix (λ + η)A")
display(Latex(latex((2+3)*A)))
print()
print("Matrix λA + ηA")
display(Latex(latex((2*A)+(3*A))))
```

**7. λ(ηA) = (λη)A**

```python
a = Vector( [1,3,5] )
b = Vector( [7,9,11] )
c = Vector( [13,15,17] )
A = Matrix( [a,b,c] )
print("Matrix A")
display(Latex(latex(A)))
print()
print()
print("λ=2  η=3")
print("Matrix λ(ηA)")
display(Latex(latex((2*(3*A)))))
print()
print("Matrix (λη)A")
display(Latex(latex((2*3)*A)))
```
